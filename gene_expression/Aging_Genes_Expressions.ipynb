```python
import pandas as pd
```

```python
gene_file='GSE40366_series_matrix.txt'
```

```python
df=pd.read_table(gene_file)
df.shape
```

```python
dft=df.transpose()
nonagenarians=dft[dft[9]=='age: nonagenarian'][0]
```

```python
dfn=df.iloc[45:]
```

```python
l=list(nonagenarians.keys())
l.append('!Sample_title')
```

```python
dfnonagenarians=dfn[l]
```

```python
l.remove('!Sample_title')
```

```python
dfnonagenarians['mean_expression'] = dfnonagenarians[l].astype(float).mean(axis=1)
```

```python
dfnonagenarians.sort_values('mean_expression',ascending=False)[['!Sample_title','mean_expression']]
```

```python
genbank=pd.read_table('genbank.txt',header=None)
```

```python
genbank.columns=['!Sample_title','gen_bank','name','note']
```

```python
dfm=dfnonagenarians.merge(genbank,on='!Sample_title')
```

```python
dfm.sort_values('mean_expression',ascending=False)\
[['mean_expression','gen_bank']].to_csv('gene_expression.csv',index=False)
```
